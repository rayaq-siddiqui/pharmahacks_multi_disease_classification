```python
labels = df['disease']
labels.head()
```

```python
# we are going to categorize the labels
# there is an incorrect matching in the csv file
# in csv Disease-1 maps to Disease-0 in the info form
# in csv Disease-3 maps to Disease-1 in the info form
# in csv Disease-2 maps to Disease-2 in the info form
# in csv Healthy maps to Healthy in the info form
possible_labels = ['Disease-1', 'Disease-3', 'Disease-2', 'Healthy']

for i, label in enumerate(labels):
    if label == possible_labels[0]:
        labels[i] = 0
    elif label == possible_labels[1]:
        labels[i] = 1
    elif label == possible_labels[2]:
        labels[i] = 2
    elif label == possible_labels[3]:
        labels[i] = 3

labels.head()
```

```python
import numpy as np

mean_normalized_df = np.asarray(mean_normalized_df).astype('float32')
labels = np.asarray(labels).astype('float32')

X_train, X_test, y_train, y_test = train_test_split(
    mean_normalized_df, 
    labels, 
    test_size=0.2, 
    random_state=3
)
```

```python
from tensorflow.keras.utils import to_categorical

y_train_categorical = to_categorical(y_train, num_classes=4)
y_train_categorical.shape
```

```python
y_test_categorical = to_categorical(y_test, num_classes=4)
y_test_categorical.shape
```

```python
# compute class weights
from sklearn.utils.class_weight import compute_class_weight

class_weight = compute_class_weight(
    class_weight='balanced',
    classes = [0,1,2,3], 
    y=labels
)
print(class_weight)

sum = 0
for i in class_weight:
    sum += i
print(sum)

class_weight = class_weight / sum
print(class_weight)
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Input, Concatenate, Maximum, Dropout, LeakyReLU
from tensorflow.keras.models import Model


def seq_model():
    model = Sequential()
    model.add(Input(shape=(1094,)))
    model.add(Dense(512, activation='relu'))
    model.add(Dense(256, activation='relu'))
    model.add(Dense(128, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(4, activation='softmax'))
    return model


def dual_input_model():
    
    inp1 = Input(shape=(1094,))
    inp2 = Input(shape=(1094,))

    x2 = Dense(512, activation='relu')(inp1)
    x1 = Dense(512, activation='relu')(x1)
    x1 = Dense(256, activation='relu')(x1)
    x1 = Dense(128, activation='relu')(x1)

    x2 = Dense(512, activation='relu')(inp2)
    x2 = Dense(512, activation='relu')(x2)
    x2 = Dense(256, activation='relu')(x2)
    x2 = Dense(128, activation='relu')(x2)

    x = Maximum()([x1,x2])
    x = Dense(256, activation='relu')(x)
    x = Dense(64, activation='relu')(x)
    out = Dense(4, activation='softmax')(x)

    model = Model([inp1, inp2], out)
    return model


class ClassWeightMult(tf.keras.layers.Layer):
    def __init__(self):
        super().__init__()
        self.class_weight = class_weight

    def call(self, inputs):
        return inputs * self.class_weight


def deep_linear_network():
        
    inp = Input(shape=(1094,))
    act = 'tanh'
    alpha = -0.5

    x = Dense(2048)(inp)
    x = LeakyReLU(alpha=alpha)(x)
    x = Dense(1024)(x)
    x = LeakyReLU(alpha=alpha)(x)

    x = Dropout(0.5)(x)

    x = Dense(512)(x)
    x = LeakyReLU(alpha=alpha)(x)
    x = Dense(256)(x)
    x = LeakyReLU(alpha=alpha)(x)

    x = Dropout(0.5)(x)

    x = Dense(128)(x)
    x = LeakyReLU(alpha=alpha)(x)
    x = Dense(64)(x)
    x = LeakyReLU(alpha=alpha)(x)

    x = Dense(4, activation='softmax')(x)
    out = ClassWeightMult()(x)

    model = Model(inp, out)
    return model
```

```python
from tensorflow.keras import optimizers
import tensorflow_addons as tfa

adam = optimizers.Adam(learning_rate=1e-3, beta_1=0.9, beta_2=0.999, epsilon=1e-07, decay=0.0)
model.compile(
    optimizer=adam, 
    loss=tf.keras.losses.CategoricalCrossentropy(),
    # loss=tfa.losses.WeightedKappaLoss(
    #     num_classes=4,
    # ),
    metrics=[
        tfa.metrics.CohenKappa(num_classes=4, weightage='quadratic'),
        tfa.metrics.F1Score(num_classes=4),
        'accuracy'
    ]
)
```

```python
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    verbose=1,
    min_lr=1e-8,
)

checkpoint = tf.keras.callbacks.ModelCheckpoint(
    'models/checkpoint',
    monitor='val_cohen_kappa',
    verbose=1,
    save_best_only=True,
    mode='max',
    save_weights_only=False
)


model.fit(
    x=X_train,
    y=y_train_categorical,
    batch_size=16,
    epochs=15,
    verbose=1,
    validation_data=(X_test, y_test_categorical),
    shuffle=True,
    callbacks=[
        reduce_lr,
        checkpoint
    ]
)
```

```python
from sklearn.metrics import cohen_kappa_score, f1_score

y_prob = model.predict(X_test) 
y_classes = y_prob.argmax(axis=-1)

ck_score = cohen_kappa_score(
    y_test,
    y_classes,
    weights='quadratic'
)

f1_score = f1_score(
    y_test,
    y_classes,
    labels=[0,1,2,3],
    average='weighted',
)

print(
    'cohen kappa score:', ck_score,
    '\nf1_score:', f1_score
)
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Input, Concatenate, Maximum, Dropout, LeakyReLU
from tensorflow.keras.models import Model

class ClassWeightMult(tf.keras.layers.Layer):
    def __init__(self):
        super().__init__()
        self.class_weight = class_weight

    def call(self, inputs):
        return inputs * self.class_weight


def xtanh(x):
    return x * tf.keras.activations.tanh(x)


def xsigmoid(x):
    return tf.keras.activations.swish(x)


def branchy_linear_network():
        
    inp = Input(shape=(1094,))
    swish = tf.keras.activations.swish
    x = Dense(2048)(inp)
    x = LeakyReLU(alpha=0.5)(x)
    x = Dense(2048)(x)
    x = LeakyReLU(alpha=0.5)(x)
    x = Dense(2048)(x)
    x = LeakyReLU(alpha=0.5)(x)
    x = Dropout(0.5)(x)

    x2 = Dense(512)(x)
    x2 = xtanh(x2)
    x3 = Dense(512)(x)
    x3 = swish(x3)

    x2 = Dense(512)(x3)
    x3 = Dense(512)(x2)
    x2 = xtanh(x2)
    x3 = swish(x3)

    x3 = Dense(512)(x2)
    x2 = Dense(512)(x3)
    x2 = xtanh(x2)
    x3 = swish(x3)

    x = Maximum()([x2,x3])
    x = Dropout(0.5)(x)

    x = Dense(128)(x)
    x = LeakyReLU(alpha=0.5)(x)
    x = Dense(128)(x)
    x = LeakyReLU(alpha=0.5)(x)
    x = Dense(128)(x)
    x = LeakyReLU(alpha=0.5)(x)

    x = Dropout(0.5)(x)

    x = Dense(4, activation='softmax')(x)
    out = ClassWeightMult()(x)

    model = Model(inp, out)
    return model


def simple_branchy_linear_network():
        
    inp = Input(shape=(1094,))
    x = Dense(2048)(inp)
    x = LeakyReLU(alpha=0.5)(x)
    x = Dense(2048)(x)
    x = LeakyReLU(alpha=0.5)(x)
    x = Dropout(0.5)(x)

    x1 = Dense(1024)(x)
    x1 = xtanh(x1)
    x1 = Dense(1024)(x1)
    x1 = xtanh(x1)
    x1 = Dense(128)(x1)
    x1 = xtanh(x1)
    x1 = Dropout(0.5)(x1)

    x2 = Dense(1024)(x)
    x2 = xsigmoid(x2)
    x2 = Dense(512)(x2)
    x2 = xsigmoid(x2)
    x2 = Dense(128)(x2)
    x2 = xsigmoid(x2)
    x2 = Dropout(0.5)(x2)

    x = Concatenate()([x1, x2])

    x = Dense(128)(x)
    x = LeakyReLU(alpha=0.5)(x)
    x = Dense(32)(x)
    x = LeakyReLU(alpha=0.5)(x)

    x = Dropout(0.5)(x)

    x = Dense(4, activation='softmax')(x)
    out = ClassWeightMult()(x)

    model = Model(inp, out)
    return model
```

```python
from tensorflow.keras.utils import plot_model

model = simple_branchy_linear_network()
model.summary()
plot_model(model, 'branchy_linear_network.png', show_shapes=True)
```

```python
from tensorflow.keras import optimizers
import tensorflow_addons as tfa

adam = optimizers.Adam(learning_rate=1e-4, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
model.compile(
    optimizer=adam, 
    loss=tf.keras.losses.CategoricalCrossentropy(),
    metrics=[
        tfa.metrics.CohenKappa(num_classes=4, weightage='quadratic'),
        tfa.metrics.F1Score(num_classes=4),
        'accuracy'
    ]
)
```

```python
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    verbose=1,
    min_lr=1e-8,
)

checkpoint = tf.keras.callbacks.ModelCheckpoint(
    'models/checkpoint',
    monitor='val_cohen_kappa',
    verbose=1,
    save_best_only=True,
    mode='max',
    save_weights_only=False
)

model.fit(
    x=X_train,
    y=y_train_categorical,
    batch_size=16,
    epochs=12,
    verbose=1,
    validation_data=(X_test, y_test_categorical),
    shuffle=True,
    callbacks=[
        reduce_lr,
        checkpoint
    ]
)
```

```python
from sklearn.metrics import cohen_kappa_score, f1_score

# load the best model
model.load_weights('models/checkpoint')

y_prob = model.predict(X_test) 
y_classes = y_prob.argmax(axis=-1)

ck_score = cohen_kappa_score(
    y_test,
    y_classes,
    weights='quadratic'
)

f1_score = f1_score(
    y_test,
    y_classes,
    labels=[0,1,2,3],
    average='weighted',
)

print(
    'cohen kappa score:', ck_score,
    '\nf1_score:', f1_score
)
```
